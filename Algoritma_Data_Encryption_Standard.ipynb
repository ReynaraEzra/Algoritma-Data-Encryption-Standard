Meskipun panjang kunci dari DES adalah 64-bit, tetapi kunci efektif hanya berukuran 56 bit saja, karena 8 dari 64 bit kunci tidak digunakan oleh algoritma enkripsi (berfungsi sebagai bit cek saja).

```python
def hex2bits(text):
    #fungsi convert hexadecimal ke binary digits
    text = list(text)
    bits = str()
    for char in text:
        char_bits = '{0:04b}'.format(int(char, 16))
        bits += char_bits
    return bits

def dec2bits(decimal):
    #fungsi convert decimal ke binary digits
    decimal = str(decimal)
    return '{0:04b}'.format(int(decimal, 10))

def bits2dec(bits):
    #fungsi convert binary digits ke decimal
    bits = str(bits)
    return int(bits, 2)

def bits2hex(bits):
    #fungsi convert binary digits ke hexadecimal
    bits = str(bits)
    return '{0:x}'.format(int(bits,2))

def permutation(pt, ip):
    # fungsi permutation block
    # pt: binary digits
    # ip: permutation block 
    permuted = [list(pt)[i-1] for i in ip]
    return ''.join(permuted)

def rotate(bits, shift = 1):
    # fungsi left shift 
    return bits[shift:]+bits[:shift]

def split_half(bits):
    # split binary digits menjadi dua bagian sama besar
    s = int(len(bits)/2)
    L = bits[:s]
    R = bits[s:]
    return L, R

def xor(Left, Right):
    # fungsi xor dari dua string yang hanya berisi nol atau satu
    L_int = [int(i) for i in list(Left)]
    R_int = [int(i) for i in list(Right)]
    xor_result = str()
    for i in range(len(L_int)):
        b = str(L_int[i]^R_int[i])
        xor_result += b
    return xor_result

def generate_B_partition(bits, n_partition = 8):
    # fungsi untuk membagi binary digits menjadi n_partition sama besar
    B_list = []
    length = len(bits)
    r = length//n_partition
    
    for i in range(n_partition):
        b_n = bits[(i*r):(i+1)*r]
        B_list.append(b_n)
    return B_list

def s_func(B, Sbox):
    # fungsi transformasi dari 6-bits ke 4-bits
    row = B[0]+B[-1]
    col = B[1:len(B)-1]
    
    row_dec = bits2dec(row)
    col_dec = bits2dec(col)
    
    idx = ((row_dec)*16 + (col_dec))
    dec = Sbox[idx]
    out = dec2bits(dec)
    return out

def f(Right, Subkey):
    # fungsi f yang digunakan pada setiap round
    result = str()
    
    Right_expanded = permutation(Right, E) #Expansion Permutation
    res = xor(Subkey, Right_expanded) #XOR
    B = generate_B_partition(res, 8) #Membagi menjadi 8 partisi (6-bit)
    
    #Apply S-boxes untuk masing-masing partisi
    for i in range(len(B)): 
        result += s_func(B[i], Sboxes[i])
    #Permutasi oleh tabel P 
    result = permutation(result, P)
    return result

def generate_subkeys(n_rounds, key):
    # fungsi pembangkit subkeys dari key sebanyak n_rounds
    C_list = []
    D_list = []
    K_list = []
    
    K = hex2bits(key)
    K_plus = permutation(K, PC1)
    
    C_n, D_n = split_half(K_plus)
    C_list.append(C_n)
    D_list.append(D_n)
    
    #Generate C_n, D_n, 1<=n<=n_rounds
    for i in range(n_rounds):
        if (i == 0) or (i == 1) or (i == 8) or (i == 15):
            C_n = rotate(C_n, 1)
            D_n = rotate(D_n, 1)
        else:
            C_n = rotate(C_n, 2)
            D_n = rotate(D_n, 2)
        C_list.append(C_n)
        D_list.append(D_n)
        
        #Generate K_n, 1<=n<=n_rounds
        concat = C_n + D_n
        K_n = permutation(concat, PC2)
        K_list.append(K_n)
        
    return K_list

def des(plaintext, key, decrypt = False):
    # fungsi utama DES, decrypt = True untuk melakukan dekripsi
    n_rounds = 16
    R_list = []
    L_list = []

    PT = hex2bits(plaintext)
    PT = permutation(PT, IP)
    L_n, R_n = split_half(PT)
    L_list.append(L_n)
    R_list.append(R_n)
    
    subkeys = generate_subkeys(n_rounds = n_rounds, key = key)
    for r in range(n_rounds):
        if decrypt:
            k = n_rounds-r-1
        else:
            k = r
            
        L_n = R_n
        R_n = xor(L_list[r], f(R_list[r], subkeys[k]))
        
        L_list.append(L_n)
        R_list.append(R_n)
    
    RL = R_n + L_n
    ip_inv_bits = permutation(RL, IP_INV)

    ciphertext = bits2hex(ip_inv_bits).zfill(16)

    return ciphertext 
```
